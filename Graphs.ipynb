### **Definition**

This part shows important graph of the data obtained in the codes. Mainly the performance of each architecture in terms of accuracy over the CIFAR-100 dataset and the amount of parameters and FLOPs used.

#### **Libraries**

```python
import matplotlib.pyplot as plt
```

#### **Graphs for the parameters and FLOPs calculation**

The information present in this section is extracted from the Parameters.ipynb.

```python
# Data
architectures = [
    "AlexNet", "VGG16", "VGG19", "ResNet50", "ResNet101", "ResNet152",
    "InceptionV3", "EfficientNetB0", "EfficientNetB1", "EfficientNetB2",
    "EfficientNetB3", "EfficientNetB4", "EfficientNetB5", "EfficientNetB6",
    "EfficientNetB7"
]
```

```python
flops = [
    0.71, 15.47, 19.63, 4.13, 7.87, 11.60, 2.86, 0.42, 0.61, 0.70,
    1.02, 1.58, 2.46, 3.49, 5.35
]

parameters = [
    61.10, 138.36, 143.67, 25.56, 44.55, 60.19, 23.83, 5.29, 7.79, 9.11,
    12.23, 19.34, 30.39, 43.04, 66.35
]
```

```python
# Function to add value labels
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i] + max(y) * 0.01, f'{y[i]:.2f}', ha='center', va='bottom')
```

```python
# Create bar plot for FLOPs
plt.figure(figsize=(12, 6))
plt.bar(architectures, flops, color='skyblue')
plt.xlabel('Architecture')
plt.ylabel('FLOPs (G)')
plt.title('FLOPs for Different Architectures')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)
add_labels(architectures, flops)
plt.tight_layout()
plt.show()
```

```python
# Create bar plot for Parameters
plt.figure(figsize=(12, 6))
plt.bar(architectures, parameters, color='lightgreen')
plt.xlabel('Architecture')
plt.ylabel('Parameters (M)')
plt.title('Parameters for Different Architectures')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)
add_labels(architectures, parameters)
plt.tight_layout()
plt.show()
```
